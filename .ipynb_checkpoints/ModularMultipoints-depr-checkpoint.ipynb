# Computing Isogenies at Singular Points of the Modular Polynomial.

## Overview
This notebook contains a collection of functions for computing isogenies between elliptic curves j1, j2 over Fq connected by multiple distinct separable l-isogenies. 

Based on the paper "Computing Isogenies at Singular points of the Modular Polynomial" by William E. Mahaney and Travis Morrison available at
https://arxiv.org/abs/2402.02038.

This notebook is draft work, for the full implementation of our algorithm see main.sage. 







```python


#Helper Functions 




#Main Functions

#Intake a pair of j-invariants j1 and j2 with m distinct l-isogenies from j1 to j2. This functions returns the models (Atidle1, Btilde1), ..., 
    #(Atildem, Btildem)$ over Fqbar for Ej2 which admit the m distinct normalized l-isogenies from j1 to j2. 
    #Requires p=Char(Fq) to be 5 or more. If 
def isogenous_curves(j1, j2, ell, mult, model = None):
    """
    Parameters
    ----------
    j1 : Element of a finite field Fq
        The j-invariant of the first elliptic curve E1.
    j2 : Element of a finite field Fq
        The j-invariant of the second elliptic curve E2.
    ell : int
        A prime integer coprime to the characteristic of the field
    mult : int
        Multiplicity of the singularity
    model : (A, B) tuple of Elements of a finite field or None.
        Optional parameter that lets the user give E1 a specific short Weierstrass model. 

    Returns
    -------
    codomain_models : list of models (Atilde, Btilde) for elliptic curves over Fqbar of j invariant j2 which 
        admit normalized l-isogenies from Ej1 to E_{Atilde, Btilde}
        

    Raises
    ------
    ValueError
        If ell is not prime or not coprime to the characteristic of the field.
        If model=(A,B) is not a valid short Weierstrass model.
        If Fq has characteristic 2 or 3.
        
        

    Notes
    -----
    Uses SageMath's built-in EllipticCurve_from_j method to construct elliptic curves when model is not given. These models are coerced 
        to short Weierstrass models.
    Uses SageMath's built-in classical_modular_polynomial(l) method to load the modular polynomials.
    """
    #k should be a number field, C, or a finite field. 
    k = j1.parent(); j2 = k(j2); #if j1 and j2 don't live in the same field or cant be coerced into the same field get an error here
    p = k.characteristic();
    
    #some common errors
    if p == 2 or p == 3:
        raise ValueError(f"Parameter p={p} must be a prime greater than 3.");
    
    if j1 == k(0) or j1 == k(1728) or j2 == k(0) or j2 == k(1728):
        #if j1, j2 = 0, 1728 then our algorithm doesn't work as we would encounter division by zero later
        raise ValueError(f"Parameters j1={j1}, j2={j2} can not be 0 or 1728");
        
    if not is_prime(ell):
        #this algorithm should work fine if l is replaced with a composite n coprime to p. 
        raise ValueError(f"Parameter l={ell} must be prime.");

    if p > 0: #don't need to check this if field is characteristic zero
        if gcd(ell, p) != 1:
            #this algorithm only works for separable isogenies.
            raise ValueError(f"Parameter l={ell} must be coprime to the characteristic p={p}.");
    if model == None:
        E1 = EllipticCurve_from_j(j1);
        E1=E1.short_weierstrass_model(); 
    else:
        #model is expected as (A, B) such that y^2 = x^3 + Ax + B has j-invariant 
        #get an error here if the model isn't right.
        E1 = EllipticCurve(model[0].parent(), model);

    #if ell == 2 use Velu instead. 
    if ell == 2: 
        return multiple_isogenies_2(j1,j2, ell, model)
    
    #now the main body
    
    #load the lth modular polynomial and its partial derivatives
    R.<X,Y>=k[]; #get the polynomial rink k[X,Y] and place Phi_l in that ring
    Phi = classical_modular_polynomial(ell); Phi=R(Phi); #put Phi in the common form 
    derivs = matrix(k, mult+1) #we compute the matrix of weight m or less partial derivaites of phil with respect to X, Y
    #This can be sped up with a smarter strategy or using instantiated modular polynomials. Using the fiber polynomial relationship 
        #lets one use half as many partial derivatives. 
    for u in range(0, mult+1):
        for v in range(0, mult+1):
            Phideriv = derivative(Phi, X, u, Y, v) 
            derivs[u,v] = Phideriv(j1,j2)

    #Now we compute the fiber polynomials
    A=E1.a4(); B=E1.a6() #E1:y^2=x^3+Ax+B
    j1prime = 18*B/A*j1
    S.<t> = k[]
    fiber_poly =0
    for u in range(0, mult+1):
        fiber_poly += binomial(mult,u) * ell^(mult-u) * j1prime^u * derivs[u, mult-u] * t^(mult-u)
    #get the roots of the fiber polynomial and then return the associated models. Note the polynomial need not split over k
    roots = [r[0] for r in fiber_poly.roots()] 
    models = []
    for r in roots:
        Atilde = ( -ell^4 * r^2)/(48*j2*(j2-1728));
        Btilde = (-ell^6 * r^3)/(864*j2^2*(j2-1728));
        models.append((Atilde, Btilde))
    return E1, models

#Intake a pair of j-invariants j1 and j2 with mult distinct ell-isogenies from j1 to j2. This functions returns the models (Atidle1, Btilde1), ..., 
    #(Atildem, Btildem)$ over Fqbar for Ej2 which admit the m distinct normalized l-isogenies from j1 to j2 and the abscissas of the 
    #isogenies p1, ..., pm. 
    #Requires p=Char(Fq) to be 5 or more. 
def isogenous_curves_abscissa(j1, j2, ell, mult, model = None):
    """
    Parameters
    ----------
    j1 : Element of a finite field Fq
        The j-invariant of the first elliptic curve E1.
    j2 : Element of a finite field Fq
        The j-invariant of the second elliptic curve E2.
    ell : int
        A prime integer coprime to the characteristic of the field
    mult : int
        Multiplicity of the singularity
    model : (A, B) tuple of Elements of a finite field or None.
        Optional parameter that lets the user give E1 a specific short Weierstrass model. 

    Returns
    -------
    codomain_models : list of models (Atilde, Btilde) for elliptic curves over Fqbar of j invariant j2 which 
        admit normalized l-isogenies from Ej1 to E_{Atilde, Btilde}
        

    Raises
    ------
    ValueError
        If ell is not prime or not coprime to the characteristic of the field.
        If model=(A,B) is not a valid short Weierstrass model.
        If Fq has characteristic 2 or 3.
        
    Notes
    -----
    Uses SageMath's built-in EllipticCurve_from_j method to construct elliptic curves when model is not given. 
        Follows up by forcing into short Weierstrass model.
    Uses SageMath's built-in classical_modular_polynomial(ell) method to load the modular polynomials.
    
    """
    #k should be a number field, C, or a finite field. 
    k = j1.parent(); j2 = k(j2); #if j1 and j2 don't live in the same field or cant be coerced into the same field get an error here
    p = k.characteristic();
    
    #some common errors
    if p == 2 or p == 3:
        raise ValueError(f"Parameter p={p} must be a prime greater than 3.");
    
    if j1 == k(0) or j1 == k(1728) or j2 == k(0) or j2 == k(1728):
        #if j1, j2 = 0, 1728 then our algorithm doesn't work as we would encounter division by zero later
        raise ValueError(f"Parameters j1={j1}, j2={j2} can not be 0 or 1728");
        
    if not is_prime(ell):
        #this algorithm should work fine if l is replaced with a composite n coprime to p. 
        raise ValueError(f"Parameter l={ell} must be prime.");

    if p > 0: #don't need to check this if field is characteristic zero
        if gcd(ell, p) != 1:
            #this algorithm only works for separable isogenies.
            raise ValueError(f"Parameter l={ell} must be coprime to the characteristic p={p}.");

    if model == None:
        E1 = EllipticCurve_from_j(j1);
        E1=E1.short_weierstrass_model(); 
    else:
        #model is expected as (A, B) such that y^2 = x^3 + Ax + B has j-invariant 
        #get an error here if the model isn't right.
        E1 = EllipticCurve(model[0].parent(), model);

    
    
    #now the main body
    
    #load the lth modular polynomial and its partial derivatives
    R.<X,Y>=k[]; #get the polynomial rink k[X,Y] and place Phi_l in that ring
    Phi = classical_modular_polynomial(ell); Phi = R(Phi)
    #we compute the matrix of weight mult+1 or less partial derivaites of phil with respect to X, Y
    derivs = matrix(k, mult+2) 
    #This can be sped up with a smarter strategy or using instantiated modular polynomials.
    for u in range(0, mult+2):
        for v in range(0, mult+2):
            Phideriv = derivative(Phi, X, u, Y, v) 
            derivs[u,v] = Phideriv(j1,j2)

    #Now we compute the fiber polynomials
    A=E1.a4(); B=E1.a6() #E1:y^2=x^3+Ax+B
    j1prime = 18*B/A*j1
    S.<t> = k[]
    fiber_poly =0
    for u in range(0, mult+1):
        fiber_poly += binomial(mult,u) * ell^(mult-u) * j1prime^u * derivs[u, mult-u] * t^(mult-u)
    #get the roots of the fiber polynomial and then return the associated models
    roots = [r[0] for r in fiber_poly.roots()] 
    codomain_data = []
    for r in roots:
        Atilde = ( -ell^4 * r^2)/(48*j2*(j2-1728));
        Btilde = (-ell^6 * r^3)/(864*j2^2*(j2-1728));
        #get the abscissa
        #first we need the mult+1 fiber function evaluated at r=j~'. 
        fiber_function_multplus1 = 0
        for u in range(0, mult+2):
            fiber_function_multplus1 += binomial(mult+1, u) * ell^(mult+1-u) * j1prime^u * derivs[u, mult+1-u] * r^(mult+1-u)
        #now we take the derivative of the fiber polynomial evaluated at our fixed model Atilde, Btilde
        fiber_poly_ddt = derivative(fiber_poly, t, 1); fiber_poly_ddt = fiber_poly_ddt(r);
        if fiber_poly_ddt == 0:
        #If we found a root corresponding to multiple models we need to abort now because our later abscissa formula
            #will involve division by 0
            codomain_data.append([[Atilde, Btilde], None])
            continue
        #now compute the difference of logarithmic derivatives given by our formula
        log_deriv_diff = mult/r * fiber_function_multplus1/( binomial(mult+1, mult-1) * fiber_poly_ddt );
        #now use Drew's formula to recover the absicca
        m1 = 18*B/A; n1 = j1prime/(1728-j1)
        m2 = r/j2; n2 = r/(1728-j2) 
        abscissa = ell * (log_deriv_diff/2 + (n1 - ell*n2)/4 + (ell*m2 - m1)/3 )
        model = [Atilde, Btilde]; 
        codomain_data.append([model, abscissa])
    return E1, codomain_data

    

def multiple_2_isogenies(j1, j2, ell, model=None):
    #intakes a pair of j-invariants j1 and j2 that are 2-isogenous and returns the isogenies between them with codomain ~= Ej2
    k = j1.parent(); j2 = k(j2); #if j1 and j2 don't live in the same field or cant be coerced into the same field get an error here
    p = k.characteristic();
    
    #some common errors
    if p == 2 or p == 3:
        raise ValueError(f"Parameter p={p} must be a prime greater than 3.");
    
    if j1 == k(0) or j1 == k(1728) or j2 == k(0) or j2 == k(1728):
        #if j1, j2 = 0, 1728 then our algorithm doesn't work as we would encounter division by zero later
        raise ValueError(f"Parameters j1={j1}, j2={j2} can not be 0 or 1728");
        
    if not is_prime(ell):
        #this algorithm should work fine if ell is replaced with a composite n coprime to p. 
        raise ValueError(f"Parameter l={ell} must be prime.");

    if ell > 2: 
        #this algorithm is only designed for ell == 2 
        raise ValueError(f"Parameter l={ell} must be 2.");

    if model == None:
        E1 = EllipticCurve_from_j(j1);
        E1 = E1.short_weierstrass_model();
    else:
        #model is expected as (A, B) such that y^2 = x^3 + Ax + B has j-invariant 
        #get an error here if the model isn't right.
        E1 = EllipticCurve(model[0].parent(), model);

    #get the isogenies from Velu by factoring the right hand side of the short Weierstrass model for E1, keep those that 
        #map to Ej2
    #outputs will be base changed to the splitting field of f
    A = E1.a4(); B = E1.a6(); #E1:y^2=x^3 + Ax + B
    k = A.parent();
    R.<t>=k[];
    f=t^3 + A*t+ B; 
    L.<w> = f.splitting_field(); S=R.base_extend(L);
    f = S(f); E1=E1.base_extend(L)
    roots = [r[0] for r in f.roots()]
    maps = [EllipticCurveIsogeny(E1, E1((r,0)), None, 2) for r in roots];
    out = []
    for map in maps:
        if map.codomain().j_invariant() == L(j2):
            out.append(map.codomain().j_invariant())
    return E1, out
        

def BMSS(E1, ell, codomain_models, abscissas):
    #intakes an elliptic curves E1 and models (A1, B1), ..., (Am, Bm) admitting normalized ell-isogenies from E1 to E_{Ai, Bi} along with 
        #the abscissas of the isogenies and returns the m distinct normalized ell-isogenies phi1, .., phim from phii: E1 -> E_{Ai, Bi}
    #error codes 

    if ell == 2 or ell == 3:
        j1 = E1.j_invariant();
        A = E1.a4(); B = E1.a6()
        m = len(codomain_models)
        return "not done"
    return "not done"

#Tests
def classical_abscissa(j1,j2,ell):
    #intakes a pair of ell-isogenous j-invariants with exactly one isogeny between them and returns the abcissa of the isogeny
    # based on pseudocode of Sutherland from Steven Galbraith's ``Mathematics of Public Key Cryptography,''
        #Chapter 25, pg555 https://www.math.auckland.ac.nz/~sgal018/crypto-book/ch25.pdf
    K=j1.parent();
    E1=EllipticCurve_from_j(j1);
    A=E1.a4(); B = E1.a6();
    R.<X,Y>=K[]
    Phi = classical_modular_polynomial(ell); Phi = R(Phi)
    derivs = matrix(K, 3)
    for u in range(0, 3):
        for v in range(0,3):
            deriv = derivative(Phi, X, u, Y, v)
            derivs[u,v] = deriv(j1,j2)
    m = 18*B/A
    j1prime = m * j1
    n = j1prime/(1728-j1)
    j2prime=(-j1prime*derivs[1,0] )/(ell*derivs[0,1])
    m2 = j2prime/j2
    n2 = j2prime/(1728-j2)
    Atilde = ell^4 * m2 * n2 / 48
    Btilde = ell^6 * m2^2 * n2 / 864
    model = [Atilde, Btilde]
    log_deriv_diff =  -(j1prime^2*derivs[2,0] + 2*ell*j1prime*j2prime*derivs[1,1] + ell^2*j2prime^2*derivs[0,2]) / (j1prime*derivs[1,0])
    rho1 = ell*(log_deriv_diff/2 + (n-ell*n2)/4 + (ell*m2-m)/3)
    return EllipticCurveIsogeny(E1,None,EllipticCurve(model),ell), rho1
    
def single_abscissa_test():
    p = 137; ell = 5; k.<w> = GF(p^2); j1 = K(22); j2 = 43*w+130; phi, rho1 = classical_abscissa(j1,j2, ell);
    print(phi.kernel_polynomial())
    print(-rho1/2)

def double_abscissa_test():
    p = 137; ell = 13; k.<w> = GF(p^2); j1 = k(22); j2 = k(78);
    E1, codomain_data = isogenous_curves_abscissa(j1,j2,ell, 2)
    out=[]
    for dat in codomain_data:
        model = dat[0]; abscissa = dat[1]
        Etilde = EllipticCurve(k, model)
        isog = EllipticCurveIsogeny(E1, None, Etilde, ell)
        psi = isog.kernel_polynomial()
        out.append([Etilde, isog, psi[(ell-1)/2-1], -abscissa/2])
    return out

def triple_abscissa_test():
    #demonstrates if isogenous_curves_abscissa works on a selected example
    p=137; ell = 7; k.<w> = GF(p^2); j1 = 43*w+130; j2 = 94*w+114;
    E1, codomain_data = isogenous_curves_abscissa(j1,j2,ell, 3)
    out = []
    for dat in codomain_data:
        model = dat[0]; abscissa = dat[1]
        Etilde = EllipticCurve(k, model)
        isog = EllipticCurveIsogeny(E1, None, Etilde, ell)
        psi = isog.kernel_polynomial()
        out.append([Etilde, isog, psi[(ell-1)/2-1], -abscissa/2])
    return out

def quad_abscissa_test():
    p=137; ell = 7; k.<w> = GF(p^2); j1 = 17*w+135; j2 = 120*w+100;
    E1, codomain_data = isogenous_curves_abscissa(j1,j2,ell,4)
    out=[]
    for dat in codomain_data:
        model = dat[0]; abscissa = dat[1]
        Etilde = EllipticCurve(k, model)
        isog = EllipticCurveIsogeny(E1, None, Etilde, ell)
        psi = isog.kernel_polynomial()
        out.append([Etilde, isog, psi[(ell-1)/2-1], -abscissa/2])
    return out


#p=137; ell = 5; k.<w> = GF(p^2); j1 = 43*w+130; j2 = 120*w+100;
#isogenous_curves_abscissa(j1,j2,ell,2)

#singleabstest()
#out = double_abscissa_test(); out
#out = triple_abscissa_test(); out
#out = quad_abscissa_test(); out
```
